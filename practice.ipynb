```python
n = int(input('Enter number - '))
if n>1:
  for i in range(2,n):
    if n%i==0:
      print('Number is not prime')
      break
  else:
    print('Number is prime')
else:
  print("Invalid number")
```

```python
l = [122,2343,32,22,22922]
largest = l[0]
index = 1

while index<len(l):
  if l[index]>largest:
    largest = l[index]
  index+=1

print(largest)
```

```python
price = [122,33,22,122,11231,1231]
import numpy as np
a = np.array(price)
print(a)

a = np.sort(a)
print(a)

al = a*2
print(al)

nl = np.multiply(a[a%2==1],0.9)
print(nl)
```

```python
mean = np.mean(price)
print(mean)

median = np.median(price)
print(median)

mode = 0
maxcount = 0

for i in price:
  count = price.count(i)
  if count>maxcount:
    maxcount = count
    mode = i

print(mode)
```
